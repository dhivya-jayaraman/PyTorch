```
import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
from sklearn import datasets
```

```
n_pts = 100
centers = [[-0.5, 0.5], [0.5,-0.5]]
X, y = datasets.make_blobs(n_samples = n_pts, random_state = 123, centers = centers, cluster_std = 0.4) #import a dataset, a cluster of data points which are surrounding a central point
x_data = torch.Tensor(X) #numpy to tensor
y_data = torch.Tensor(y.reshape(100,1)) #numpy to tensor
```

```
def scatter_plot():
  plt.scatter(X[y==0, 0], X[y==0, 1]) #plots all data points who's x and y points have output(y)=0
  plt.scatter(X[y==1, 0], X[y==1, 1]) #plots all data points who's x and y points have output(y)=1
```

```
scatter_plot()
```

```
class Model(nn.Module):
  def __init__(self, input_size, output_size):
    super().__init__()
    self.linear = nn.Linear(input_size, output_size)
  def forward(self, x):
    pred = torch.sigmoid(self.linear(x)) #find the value of x and converts to probability(using sigmoid function)
    return pred
  def predict(self, x):
    pred = self.forward(x)
    if pred >= 0.5:
      return 1
    else:
      return 0
    
```

```
torch.manual_seed(2)
model = Model(2,1) #model
[w, b] = model.parameters()
w1, w2 = w.view(2)
def get_params():
  return (w1.item(), w2.item(), b[0].item())
  
```

```
def plot_fit(title):
  plt.title = title
  w1, w2, b = get_params()
  x1 = np.array([-2.0,2.0])
  x2 = (w1*x1 + b) / (-w2)
  plt.plot(x1,x2,'r')
  scatter_plot()
```

```
plot_fit('initial')
```

```
criterion = nn.BCELoss() #Binary Cross Entropy for error calculation
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01) #stochastic Gradient Descent
```

```
epochs = 1000
losses = []
for i in range(epochs):
  y_pred = model.forward(x_data)
  loss = criterion(y_pred, y_data) #calculates the error
  print ('epoch: ',i,'loss: ',loss.item())
  losses.append(loss.item())
  optimizer.zero_grad() #reset the optimizer value because it will accumulate
  loss.backward() #calculate gradient descent 
  optimizer.step() #update the model parameters
```

```
plt.plot(range(epochs), losses)
plt.xlabel('epochs')
plt.ylabel('loss')
```

```
plot_fit('Trained data')
```

```
point1 = torch.Tensor([-1.9,-1.8])
point2 = torch.Tensor([0.5,0.1])
plt.plot(point1.numpy()[0], point1.numpy()[1], 'ro')
plt.plot(point2.numpy()[0], point2.numpy()[1], 'ko')
print ("red point in class {} ".format(model.predict(point1)))
print ("black point in class {} ".format(model.predict(point2)))
plot_fit('Trained data')
```
